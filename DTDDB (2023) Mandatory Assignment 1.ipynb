## Task 9 (2 points)

__Timothy__: _I know I know, I've bugged you enough, but there's one last thing you could do for me. It would be super convenient for me if I could just plug in these four price categories into some type of function maybe and get info about the number of listings inside that category as well as the min., max., and avg. prices in these categories. Can you provide me with such a function?_

Hint: The function should take a price category as a parameter and return the info requested by Timothy. For the A+ answer, you also might want to handle cases in which an invalid price category was given.

```python
# Function to analyse info about number of listings

def get_price_category_info(category):
    # Calculate quartiles
    quartiles = data['price'].quantile([0.25, 0.5, 0.75])

    # Define the price categories
    budget = data[data['price'] <= quartiles.iloc[0]]
    moderate = data[(data['price'] > quartiles.iloc[0]) & (data['price'] <= quartiles.iloc[1])]
    premium = data[(data['price'] > quartiles.iloc[1]) & (data['price'] <= quartiles.iloc[2])]
    luxury = data[data['price'] > quartiles.iloc[2]]

    # Deciding which filtered data to be considered based on user choice.
    if category == 'budget':
        category_data = budget
    elif category == 'moderate':
        category_data = moderate
    elif category == 'premium':
        category_data = premium
    elif category == 'luxury':
        category_data = luxury
    else:
        return "Invalid category"

    # Calculate the number of listings in the category
    num_listings = len(category_data)

    # Calculate min, max, and average prices in the category
    min_price = category_data['price'].min()
    max_price = category_data['price'].max()
    avg_price = category_data['price'].mean()

    return {
        'Category': category,
        'Number of Listings': num_listings,
        'Min Price': min_price,
        'Max Price': max_price,
        'Average Price': avg_price
    }
```
